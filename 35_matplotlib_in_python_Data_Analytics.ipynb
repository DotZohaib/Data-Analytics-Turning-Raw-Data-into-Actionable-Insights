<h1>matplotlib in python</h1>

Matplotlib is a comprehensive plotting library in Python, widely used for creating static, animated, and interactive visualizations. It provides an object-oriented API for embedding plots into applications and supports various plot types, including line plots, scatter plots, histograms, bar charts, and 3D plots. 

<h1>Bar Plot in Matplotlib</h1>

```python
import matplotlib.pyplot as plt
```

```python
x = ["a", "b", "c", "d", "e"]
y = [79, 90, 61, 52, 83]

plt.bar(x, y, color=["red", "green", "blue", "yellow", "orange"])
plt.title("CodeWithZohaib")
plt.xlabel("DotZohaib")
plt.ylabel("DotIqra")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data bana rahe hain

df = pd.read_excel("multiple_linear_regression.xlsx")
df.head(5)

plt.bar(df["Experience (Years)"], df["Salary"])
plt.show()
```

<h1>Line Plot in matplotlib</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data bana rahe hain

df = pd.read_excel("multiple_linear_regression.xlsx")
df.head(5)

plt.plot(df["Experience (Years)"], df["Salary"], marker="o")
# plt.plot(df["Experience (Years)"], df["Salary"], marker = "^", color = "blue", ls ="--")
plt.show()
```

```python
import pandas as pd

x = ["2016", "2017", "2018", "2019", "2020"]
y = [79, 90, 61, 52, 83]
y1 = [39, 50, 81, 72, 43]

plt.plot(x, y, color="red", label="yearly", marker="o")
plt.plot(x, y1, color="blue", label="weak", marker="^")
plt.title("How many hours do you study?")
plt.xlabel("yearly")
plt.ylabel("weak")
plt.legend()
plt.show()
```

<h1>Scatter Ploting in matplotlib</h1>

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1, 10, 50)
y = np.random.randint(10, 100, 50)
color = np.random.randint(1, 10, 50)

plt.scatter(x, y, marker="*", cmap = "hot", c =color)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel("multiple_linear_regression.xlsx")
df.head(5)

plt.scatter(df["Experience (Years)"], df["Salary"], marker="*")
plt.show()
```

<h1>Pie Chart in matplotlib</h1>

```python
import matplotlib.pyplot as plt


brand = ["Apple", "Samsung", "Oppo", "Vivo", "Realme"]
sales = [10, 20, 30, 40, 50]


ex=[0,0,0,0,0.1]
plt.pie(sales, labels=brand, explode=ex, shadow = True, autopct="%.2f")
# plt.legend()
plt.show()
```

```python
import pandas as pd

data = pd.read_excel("multiple_linear_regression.xlsx")
data.head(5)

plt.pie(data["Salary"])
plt.show()
print(data)
```
